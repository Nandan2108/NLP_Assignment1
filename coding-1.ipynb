```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

# Download tokenizer model (run once)
nltk.download('punkt')

# Input paragraph
paragraph = (
    "Artificial intelligence is changing the world rapidly. "
    "It’s being used in healthcare, education, and business. "
    "However, many people don’t fully understand how AI works."
)

# 1. NLP Tool Tokenization (NLTK)
tool_tokens = word_tokenize(paragraph)

print("NLTK Tokenization:")
print(tool_tokens)

print("\n" + "-" * 50 + "\n")

# 2. Manual tokenization (from previous step)
manual_tokens = [
    "Artificial", "intelligence", "is", "changing", "the", "world",
    "rapidly", ".", "It", "’s", "being", "used", "in", "healthcare",
    ",", "education", ",", "and", "business", ".", "However", ",",
    "many", "people", "do", "n’t", "fully", "understand", "how",
    "AI", "works", "."
]

print("Manual Tokenization:")
print(manual_tokens)

print("\n" + "-" * 50 + "\n")

# 3. Compare tokens
tool_set = set(tool_tokens)
manual_set = set(manual_tokens)

only_in_tool = tool_set - manual_set
only_in_manual = manual_set - tool_set

print("Tokens only in NLP tool output:")
print(only_in_tool)

print("\nTokens only in manual tokenization:")
print(only_in_manual)
```
