```python
import re

# Input paragraph
paragraph = (
    "Artificial intelligence is changing the world rapidly. "
    "It’s being used in healthcare, education, and business. "
    "However, many people don’t fully understand how AI works."
)

# 1. Naïve space-based tokenization
naive_tokens = paragraph.split()

print("Naïve Tokenization:")
for token in naive_tokens:
    print(token)

print("\n" + "-"*40 + "\n")

# 2. Manually corrected tokenization
# - Separate punctuation
# - Split contractions (’s, n’t)
corrected_text = paragraph

# Separate punctuation
corrected_text = re.sub(r"([.,!?])", r" \1 ", corrected_text)

# Handle common clitics
corrected_text = corrected_text.replace("It’s", "It ’s")
corrected_text = corrected_text.replace("don’t", "do n’t")

# Remove extra spaces
corrected_text = re.sub(r"\s+", " ", corrected_text).strip()

corrected_tokens = corrected_text.split(" ")

print("Manually Corrected Tokenization:")
for token in corrected_tokens:
    print(token)
```
