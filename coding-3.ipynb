```python
from nltk.tokenize import word_tokenize
from nltk.tokenize.mwe import MWETokenizer
import nltk

# Make sure required data is downloaded
nltk.download('punkt')
nltk.download('punkt_tab')

# Input text
text = (
    "Artificial intelligence is transforming the healthcare system. "
    "Many AI startups are based in New York City."
)

# Step 1: Normal tokenization
tokens = word_tokenize(text)
print("Before MWE Tokenization:")
print(tokens)

print("\n" + "-" * 50 + "\n")

# Step 2: Define Multiword Expressions
mwe_list = [
    ("Artificial", "intelligence"),
    ("healthcare", "system"),
    ("New", "York", "City")
]

# Step 3: Apply MWE tokenizer
mwe_tokenizer = MWETokenizer(mwe_list, separator="_")
mwe_tokens = mwe_tokenizer.tokenize(tokens)

print("After MWE Tokenization:")
print(mwe_tokens)
```
